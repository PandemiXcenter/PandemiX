```python
import numpy as np
import matplotlib.pyplot as plt
# Make functions for MCMC
def logiModel(xdata,theta):
    x0 = theta[0]
    a = theta[1]
    return 1/(1+((1-x0)/x0)*np.exp(-a*xdata))


fig,ax1 = plt.subplots(tight_layout=True)

tts = np.linspace(0,30)
# yys = logiModel(tts,[0.01,0.10])
# ax1.plot(tts,yys,'k:') 
yys = logiModel(tts,[0.01,0.15])
ax1.plot(tts,yys,'k:',linewidth=3) 
yys = logiModel(tts,[0.01,0.2])
ax1.plot(tts,yys,'k:',linewidth=3) 
yys = logiModel(tts,[0.01,0.3])
ax1.plot(tts,yys,'k:',linewidth=3) 
yys = logiModel(tts,[0.01,0.4])
ax1.plot(tts,yys,'k:',linewidth=3) 
yys = logiModel(tts,[0.01,0.6])
ax1.plot(tts,yys,'k:',linewidth=3) 

ax1.set_ylabel('Andel Omikron',fontsize=20)
ax1.set_xlabel('Tid',fontsize=20)
plt.xticks([])
plt.yticks([])
plt.ylim(bottom=0,top=1)
fig.patch.set_facecolor('white')

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

plt.savefig('LogiEksempel')
```
