```python
# Define paths
rootdir_data = os.getcwd() +"/../variant data/" 
path_figs = os.getcwd() +"/../Figures/" 

# Load ACDC data
df_acdc = pd.read_csv( rootdir_data + "dataacdc.csv")
df_acdc['year_week'] = pd.to_datetime(df_acdc.year_week + '1',format='%Y-%W%w')

df_acdc = df_acdc[~(df_acdc['year_week'] < '2022-02-01')]
```

```python
list_of_loc = ["Denmark", "Portugal", "South Africa", "Germany", "United Kingdom" ]
for loc in list_of_loc:
##############################################################################################################
    #Load data from https://cov-spectrum.org/explore/Denmark/AllSamples/Past6M/variants
    BA4_cases = pd.read_csv(rootdir_data + "/"+loc + "/"+loc+"_BA4_cases.csv")
    BA4_prop = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA4_proportion_of_sequences.csv")
    BA4_abs = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA4_Abs_of_sequences.csv")

    BA5_cases = pd.read_csv(rootdir_data + "/"+loc + "/"+loc+"_BA5_cases.csv")
    BA5_prop = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA5_proportion_of_sequences.csv")
    BA5_abs = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA5_Abs_of_sequences.csv")

    #Fix datetime for dfs
    BA4_prop["date"] = pd.to_datetime(BA4_prop.date, format='%Y-%m-%d')
    BA4_cases["date"] = pd.to_datetime(BA4_cases.date, format='%Y-%m-%d')
    BA4_abs["date"] = pd.to_datetime(BA4_abs.date, format='%Y-%m-%d')

    BA5_prop["date"] = pd.to_datetime(BA5_prop.date, format='%Y-%m-%d')
    BA5_cases["date"] = pd.to_datetime(BA5_cases.date, format='%Y-%m-%d')
    BA5_abs["date"] = pd.to_datetime(BA5_abs.date, format='%Y-%m-%d')

    ############################################################################################################

    #get our world in data, data
    df_owid = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
    df_owid['date'] =  pd.to_datetime(df_owid['date'], format='%Y-%m-%d')

    #Choose loc
    df_owid_loc = df_owid.loc[df_owid["location"] == loc]

    #Cut to match dates
    start = pd.Timestamp('2022-3-1')
    mask = (df_owid_loc['date'] > start)
    df_owid_loc = df_owid_loc.loc[mask]

    ###########################################################################################################
    #Figure stuff
    fig, ax = plt.subplots(2,2, figsize=(30,12), sharex=False)

    fig.patch.set_facecolor('xkcd:off white')
    ax[0,0].set_facecolor('xkcd:off white')
    ax[0,1].set_facecolor('xkcd:off white')
    ax[1,0].set_facecolor('xkcd:off white')
    ax[1,1].set_facecolor('xkcd:off white')

    subfontsize = 16 #Size of subfigure titels
    #####################################################################################################################
    #Fig 1
    ########################################################################################################################
    #Plot data for ax[]
    ### For proportional
    # ax[0,0].plot(BA4_prop['date'], BA4_prop['estimatedCases'], label="BA4 sampels", color='b', linestyle='-', marker=".")
    # ax[0,0].fill_between(BA4_prop['date'], BA4_prop['estimatedCasesCILower'], BA4_prop["estimatedCasesCIUpper"], color='b', alpha=.2)

    # ax[0,0].plot(BA5_prop['date'], BA5_prop['estimatedCases'], label="BA5 sampels", color='g', linestyle='-', marker=".")
    # ax[0,0].fill_between(BA5_prop['date'], BA5_prop['estimatedCasesCILower'], BA5_prop["estimatedCasesCIUpper"], color='g', alpha=.2)
    # ax[0,0].set_ylabel("Proportion of all sampels")


    # ymax = max([BA4_prop['estimatedCases'].max(),BA5_prop['estimatedCases'].max()])
    # ymax = 1.1*ymax
    # ax[0,0].set_ylim([0,ymax])

    #Cases
    ax[0,0].plot(BA4_prop['date'], BA4_prop['estimatedCases'], label="BA4", color='b', linestyle='-', marker=".")
    ax[0,0].fill_between(BA4_prop['date'], BA4_prop['estimatedCasesCILower'], BA4_prop["estimatedCasesCIUpper"], color='b', alpha=.2)

    ax[0,0].plot(BA5_prop['date'], BA5_prop['estimatedCases'], label="BA5", color='g', linestyle='-', marker=".")
    ax[0,0].fill_between(BA5_prop['date'], BA5_prop['estimatedCasesCILower'], BA5_prop["estimatedCasesCIUpper"], color='g', alpha=.2)
    ax[0,0].set_ylabel("Proportion of samples")

    ax1b = ax[0,0].twinx()
    ax1b.plot(df_owid_loc['date'],df_owid_loc['new_cases_smoothed_per_million'], label="New cases per mil", color="purple",linestyle='--')
    ax1b.set_ylabel("New Cases pr mil")
    ax1b.yaxis.label.set_color("purple")


    tkw = dict(size=4, width=1.5)
    ax1b.tick_params(axis='y', colors="purple", **tkw)

    #Set time for plot, start from first of March
    ax[0,0].set_xlim([datetime.date(2022, 3, 1),df_owid_loc['date'].max()])

    ## set y max to cases!
    ymax = max([BA4_prop['estimatedCases'].max(),BA5_prop['estimatedCases'].max()])
    ymax = 1.1*ymax
    ax[0,0].set_ylim([0,ymax])
    ax[0,0].xaxis.set_major_formatter(mdates.DateFormatter('%d\n%b'))

    vac = f"{df_owid_loc['people_fully_vaccinated_per_hundred'].dropna().iloc[0]}% - {df_owid_loc['people_fully_vaccinated_per_hundred'].dropna().iloc[-1]}%"
    booster = f"{df_owid_loc['total_boosters_per_hundred'].dropna().iloc[0]}% - {df_owid_loc['total_boosters_per_hundred'].dropna().iloc[-1]}%"

    #df_owid_loc.dropna(subset=['date', 'hosp_patients_per_million'])

    ax[0,0].set_title(f"Estimated cases with linages and new cases per mil", fontsize=subfontsize)

    legend_elements = [ Line2D([0], [0],  label="New cases per mil", color="purple",linestyle='--'),
                           Line2D([0],[0],label="BA5 estimated cases", color='g', linestyle='-', marker="."),
                           Line2D([0],[0], label="BA4 estimated cases", color='b', linestyle='-', marker=".") ]

    ax[0,0].legend(handles=legend_elements, loc='upper left')
    ########################################################################################################################
    #Fig 2
    ########################################################################################################################
    #if loc != "South Africa":
    LIST = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia',
           'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece',
           'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia',
           'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands',
           'Norway', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia',
           'Spain', 'Sweden']
    if loc in LIST:

        df_loc = df_acdc.loc[df_acdc["country"] == loc] # Finde country
        df_loc = df_loc.loc[df_loc["source"] == 'GISAID'] # Find sourse

        A = df_loc.pivot_table(index="year_week", columns="variant", values="percent_variant")
        start = A.index.searchsorted(datetime.datetime(2022, 3, 1))
        A = A.iloc[start::]
        #Small variance detection. removes BA5 and BA4, then collects and adds to "others"
        B = A.loc[: , A.sum(axis=0) <= 10]
        C = A.loc[: , A.sum(axis=0) > 10]
        if "BA.4" in B:
            C["BA.4"] = B["BA.4"]
            B = B.drop("BA.4", axis=1)
        if "BA.5" in B:
            C["BA.5"] = B["BA.5"]
            B = B.drop("BA.5", axis=1)
        
        if "Other" in C:
            C["Other"] = B.sum(axis=1)+C["Other"]
        else:
            C["Other"] = B.sum(axis=1)
        C = C.cumsum(axis=1)

        ax[0,1].fill_between(C.index, C[C.columns[0]], label = C.columns[0])

        for previous, current in zip(C.columns, C.columns[1:]):
            ax[0,1].fill_between(C.index, C[previous], C[current], label = current)
        ax[0,1].legend(loc=3)

        #ax1.set_xlim([datetime.date(2022, 4, 1), C.index[-1]])
        ax[0,1].set_ylim([0,100])
        ax[0,1].set_ylabel("% of strains")




        ax_ins = inset_axes(ax[0,1],
                            width="30%", # width = 30% of parent_bbox
                            height="30%", # height : 1 inch
                            loc=1)
        plt.xticks(visible=False)
        plt.yticks(visible=True)

        ax_ins.fill_between(C.index, C[C.columns[0]], label = C.columns[0])

        for previous, current in zip(C.columns, C.columns[1:]):
            ax_ins.fill_between(C.index, C[previous], C[current], label = current)

        ax_ins.set_xlim([datetime.date(2022, 4, 1), C.index[-1]])
        ax_ins.set_ylim([95,100])
        ax[0,1].xaxis.set_major_formatter(mdates.DateFormatter('%d\n%b'))
        ax[0,1].set_title(f"linage proportion of sequenced sampels", fontsize=subfontsize)
    # fmt = mdates.DateFormatter('%Y\n Week %W')
    # ax1.xaxis.set_major_formatter(fmt)

    ########################################################################################################################
    #Fig 3
    ########################################################################################################################

    c1,c2,c3 = "xkcd:bright purple","xkcd:ultramarine","red"
    ax[1,0].plot(df_owid_loc['date'],df_owid_loc['new_deaths_smoothed_per_million'], label="New COVID-deaths", color=c3, linestyle='-')

    ax5b = ax[1,0].twinx()
    ax5b.plot(df_owid_loc.dropna(subset=['date', 'hosp_patients_per_million'])['date'], df_owid_loc.dropna(subset=['date', 'hosp_patients_per_million'])["hosp_patients_per_million"].dropna(), label="COVID-patients in hospitals", color= c2,linestyle='-')

    ax5c = ax[1,0].twinx()
    ax5c.spines.right.set_position(("axes", 1.2))
    ax5c.plot(df_owid_loc.dropna(subset=['date', 'icu_patients_per_million'])['date'],df_owid_loc.dropna(subset=['date', 'icu_patients_per_million'])['icu_patients_per_million'], label="COVID-patients in ICU", color= c1 , linestyle='-')

    ax[1,0].yaxis.label.set_color(c3)
    ax5b.yaxis.label.set_color(c2)
    ax5c.yaxis.label.set_color(c1)


    tkw = dict(size=4, width=1.2)

    ax[1,0].set_ylabel("New Covid Deaths per mil")
    ax[1,0].tick_params(axis='y', colors=c3, **tkw)

    ax5b.set_ylabel("Covid patients in hospitals per mil")
    ax5b.tick_params(axis='y', colors=c2, **tkw)

    ax5c.set_ylabel("Covid patients in ICU per mil")
    ax5c.tick_params(axis='y', colors=c1, **tkw)


    ax5d = ax[1,0].twinx()
    ax5d.fill_between(BA5_cases['date'], BA5_cases['estimatedCasesCILower'], BA5_cases["estimatedCasesCIUpper"], color='g', alpha=.2)
    ax5d.fill_between(BA4_cases['date'], BA4_cases['estimatedCasesCILower'], BA4_cases["estimatedCasesCIUpper"], color='b', alpha=.2)
    ax5d.yaxis.set_ticklabels([])

    legend_elements = [ Line2D([0], [0], color=c3, linestyle="-", label="New COVID-deaths"),
                           Line2D([0],[0], color=c2, linestyle="-",label="COVID-patients in hospitals"),
                           Line2D([0],[0], color=c1, linestyle="-",label="COVID-patients in ICU"),
                      Patch(facecolor='b', label="BA4 estimated cases", alpha=.2),
                      Patch(facecolor='g', label="BA5 estimated cases", alpha=.2)
                      ]
    ax[1,0].legend(handles=legend_elements, loc='lower left',fontsize=14) # bbox_to_anchor=(0.4,0.7)

    ax[1,0].set_xlim([datetime.date(2022, 3, 1),df_owid_loc['date'].max()])
    # ymax = max([BA4_cases['estimatedCases'].max(),BA5_cases['estimatedCases'].max()])
    # ymax = 1.1*ymax
    # ax5.set_ylim([0,ymax])
    ax[1,0].xaxis.set_major_formatter(mdates.DateFormatter('%d\n%b'))
    ax[1,0].set_title(f"Deaths, ICU and hospitalized patients with Covid", fontsize=subfontsize)


    ########################################################################################################################
    #Fig 4
    ########################################################################################################################
    BA4_abs["total sequences"] = BA4_abs["estimatedCases"]/BA4_prop["estimatedCases"]
    BA5_abs["total sequences"] = BA5_abs["estimatedCases"]/BA5_prop["estimatedCases"]
    
    
    
    ax[1,1].bar(BA4_abs["date"], BA4_abs["total sequences"],color="xkcd:darkblue", alpha=0.4)
    ax[1,1].bar(BA5_abs["date"], BA5_abs["total sequences"],color="xkcd:darkblue", alpha=0.4)
    axt2 = ax[1,1].twinx()
    axt2.plot(df_owid_loc["date"],df_owid_loc["new_tests_smoothed_per_thousand"],'-' ,color="xkcd:red")


    legend_elements = [ Line2D([0], [0], color="xkcd:red", linestyle="-", label="New tests pr capita"),
                      Patch(facecolor="xkcd:darkblue", label="Sampels sequenced", alpha=.4),
                      ]
    ax[1,1].legend(handles=legend_elements, loc='upper left',fontsize=14) # bbox_to_anchor=(0.4,0.7)


    #Set time for plot, start from first of March
    ax[1,1].set_xlim([datetime.date(2022, 3, 1),df_owid_loc['date'].max()])
    ax[1,1].set_ylabel("Sampels sequenced")
    ax[1,1].tick_params(axis='y', colors="xkcd:darkblue", **tkw)
    ax[1,1].yaxis.label.set_color("xkcd:darkblue")

    axt2.set_ylabel("New tests pr thousand")
    axt2.tick_params(axis='y', colors="xkcd:red", **tkw)
    axt2.yaxis.label.set_color("xkcd:red")

    ax[1,1].set_title(f"Total number of sampels sequenced \n and new tests done pr capita", fontsize=subfontsize)
    ax[1,1].xaxis.set_major_formatter(mdates.DateFormatter('%d\n%b'))
    ########################################################################################################################
    #Total fig
    ########################################################################################################################
    plt.subplots_adjust(wspace=0.55, hspace=0.4)
    #plt.subplot_tool()
    fig.suptitle(f"{loc} \n Vaccinated:[{vac}], Booster:[{booster}]",fontsize=20)
    sourses = "[https://ourworldindata.org]\n[https://cov-spectrum.org]\n[https://www.ecdc.europa.eu/en/publications-data/data-virus-variants-covid-19-eueea]"
    plt.annotate(sourses, (-0.4,0), (-80,-20), fontsize=12, 
             xycoords='axes fraction', textcoords='offset points', va='top')
    #plt.tight_layout()
    if saveFigures:
        fig.savefig(path_figs+f'/Variance figures/{loc} BA4 and BA5 overview',bbox_inches='tight')
    plt.close(fig)
```

```python
fig, ax = plt.subplots(figsize=(14,10))
fig.patch.set_facecolor('xkcd:off white')
ax.set_facecolor('xkcd:off white')

list_of_loc = ["Denmark", "Portugal", "South Africa", "Germany", "United Kingdom"]
time_delta_list = ['75 days', '35 days', '0 days', '70 days','65 days']
loc_colors = ["r","g","y","k","b"]
i = 0
for loc in list_of_loc:
    df_owid_loc = df_owid.loc[df_owid["location"] == loc]
    start = pd.Timestamp('2022-3-1')
    mask = (df_owid_loc['date'] > start)
    df_owid_loc = df_owid_loc.loc[mask]
    
    #Load data from https://cov-spectrum.org/explore/Denmark/AllSamples/Past6M/variants
    BA4_cases = pd.read_csv(rootdir_data + "/"+loc + "/"+loc+"_BA4_cases.csv")
    BA4_prop = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA4_proportion_of_sequences.csv")

    BA5_cases = pd.read_csv(rootdir_data + "/"+loc + "/"+loc+"_BA5_cases.csv")
    BA5_prop = pd.read_csv(rootdir_data + "/"+loc + "/"+loc + "_BA5_proportion_of_sequences.csv")

    #Fix datetime for dfs
    BA4_prop["date"] = pd.to_datetime(BA4_prop.date, format='%Y-%m-%d')
    BA4_cases["date"] = pd.to_datetime(BA4_cases.date, format='%Y-%m-%d')

    BA5_prop["date"] = pd.to_datetime(BA5_prop.date, format='%Y-%m-%d')
    BA5_cases["date"] = pd.to_datetime(BA5_cases.date, format='%Y-%m-%d')

    BA5_prop["sum"] = BA5_prop["estimatedCases"]+BA4_prop["estimatedCases"]
    BA5_cases["sum"] = BA5_cases["estimatedCases"]+BA4_cases["estimatedCases"]
    
    time_delta = time_delta_list[i]
    ax.plot(BA5_prop['date']-pd.Timedelta(time_delta), BA5_prop['sum'],
            label=f"{list_of_loc[i]} timedelta = {time_delta_list[i]}", linestyle='-', color=loc_colors[i])
    if loc == "Denmark":
        latest_data = BA5_prop['date'].iloc[-1].strftime('%Y-%m-%d')

        ax.annotate(f'{latest_data}', (mdates.date2num(BA5_prop['date'].iloc[-1]-pd.Timedelta(time_delta)), BA5_prop['sum'].iloc[-1]), xytext=(15, 15), 
                    textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

        ax.axvline(mdates.date2num(BA5_prop['date'].iloc[-1]-pd.Timedelta(time_delta)), linestyle='--', color='grey', alpha=0.4)
    i = i+1


#legend_elements = [Patch(facecolor=loc_colors[i], label=f"{list_of_loc[i]} timedelta = {time_delta_list[i]}") for i in range(4)]
ax.legend(loc='upper left',fontsize=14)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%d\n%b'))
plt.tight_layout()
ax.set_title(f" Timeshifted BA4 + BA5 proportion of sequences")
ax.set_ylabel("BA4+BA5 proportion")
ax.set_xticks([])
#Zoom
ax.set_xlim([datetime.date(2022, 3, 1),datetime.date(2022, 4, 1)])
# ymax = max([BA4_cases['estimatedCases'].max(),BA5_cases['estimatedCases'].max()])
# ymax = 1.1*ymax
# ax.set_ylim([0,ymax])
ax.set_ylim([0,0.5])

if saveFigures:
    fig.savefig(path_figs+f'/Variance figures/Time Shifted variance sums',bbox_inches='tight')
plt.close(fig)
```
