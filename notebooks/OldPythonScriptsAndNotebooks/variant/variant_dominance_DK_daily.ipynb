```python
yearweek_to_date = lambda d, weekday="-1": dt.datetime.strptime(d + weekday, "%Y-%W-%w")
#this is stolen from rasmus' notebook
# yearweek_to_date = lambda x: dt.date.fromisocalendar(int(x[:4]),int(x[5:]),1)

# Model tests
#    for fits
logistic_model = lambda t, delta, p0: 1/(1+((1-p0)/p0)*np.exp(-delta*t))
lin_model = lambda t, alpha, beta: beta + alpha*t

# Data transform
logit = lambda p: np.log(p/(1-p))
logit_err = lambda p: np.log(1/(p*(1-p)))

#Error analysis:
#    probability p changes over time as a logistic.
#    the sampling of the population at any point in time is binomial as a function of p
#    the standard std of a binomial is \sigma=sqrt(Np(1-p))
std_binom = lambda p, n: np.sqrt(p*(1-p)/n)
```

```python
path=""
filename="SSI_delta_2021_07_06.xlsx"
df_raw = pd.read_excel(path+filename)
df_raw
```

```python


# print(df.columns)
rename_dict = {
    "Pr√∏vedato":"Dato", 
    "Positive PCRtest, hvor der er foretaget Variant-PCR analyse":"numTest", 
    "Formodet deltavariant ":"numDelta", 
    "Inkonklusiv":"numInkon", 
    "Andre varianter":"numOther", 
    "Deltavariant i pct. af Variant-PCR test":"percDelta"
}
df = df_raw.rename(columns=rename_dict)
df.percDelta = pd.to_numeric(df.percDelta.astype("str").str.replace(",", ".", regex=False))
# df = df.sort_values(by='Dato').reset_index(drop=True)
df
```

```python
# not used...
#variants = ["Inconclusiv", "Andre varianter", "Formodet deltavariant"]
#variant_color = ["gray", "blue", "red"]

nu = 1./5

# v_col = {v:vc for v,vc in zip(variants, variant_color)}

# Generate what is needed, clean out what is not
df["propDelta"] = df.percDelta/100
df["propDeltaErr"] = std_binom(df.propDelta, df.numTest)
```

```python
#dataframe for saving results: 
# Logistic model: p' = delta p (1-p) => p(t) = 1/(1+((1-y0)/y0)*np.exp(-delta*t))
    # delta 
# linear model: f = \alpha x + \beta
    #\alpha is incline of 
    #\beta is the const. offset 
fit_results_df = pd.DataFrame(columns=["delta", "p0", "alpha", "beta", "alpha_transf", "beta_transf"])

#filtering for problematic datapoints...
filt_no_extremes = (df.propDelta > 0)&(df.propDelta < 1)
df_filt = df.loc[filt_no_extremes]
threshold = 5
include_filt = ((df_filt.propDelta).apply(logit_err) <= threshold)
df_filt = df_filt.loc[include_filt]
#df_filt
```

```python
#fig, ax = plt.subplots(figsize=(16,8))
#ax.errorbar(df_filt.curDates, df_filt.propDelta, yerr=df_filt.propDeltaErr)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(12, 12), tight_layout=True, sharex=True)

v = "$\\delta$-variant"

# Defining the data for plotting and fitting:
X_data = df_filt.Dato
X_data_num = (pd.to_datetime(X_data.values) - pd.to_datetime(X_data.values[0])) / np.timedelta64(1,'D')

Y_data_raw = df_filt.propDelta
Y_data_transf = Y_data_raw.apply(logit)

Y_data_raw_err = df_filt.propDeltaErr
Y_data_transf_err = Y_data_raw.apply(lambda y: abs(logit_err(y)))*Y_data_raw_err # via error propagation

## Fitting
# performing logistic fitting procedure and plotting of fit:
p0_init = [0.05, 0.003]
popt_logis, pcov_logis = curve_fit(logistic_model, X_data_num, Y_data_raw, sigma=None, p0=p0_init, maxfev=1200)
delta, p0 = popt_logis
alpha_transf, beta_transf = delta, logit(p0) #np.log(p0/(1-p0))

# performing linear fitting procedure and plotting of fit:
p0_init = None
popt_lin, pcov_lin = curve_fit(lin_model, X_data_num, Y_data_transf, sigma=Y_data_transf_err, p0=p0_init, maxfev=1200)
alpha, beta = popt_lin

params_dict = {
    "delta":delta, "p0":p0, "alpha":alpha, "beta":beta, 
    "alpha_transf":alpha_transf, "beta_transf":beta_transf
}
fit_results_df = fit_results_df.append(params_dict, ignore_index=True)

## Plotting:
plotstyle = {"marker":"_", "markersize":10, "linestyle":":"}

# performing the dataplots and fit for filtered data:
axs[0].errorbar(X_data, Y_data_raw, yerr=2*Y_data_raw_err, **plotstyle, label=f"{v}")

#Logistic fit-plot
axs[0].plot(X_data, logistic_model(X_data_num, *popt_logis), label=f"Logistic fit to {v}")
axs[0].tick_params(axis='x', labelrotation=90)
axs[0].set_ylabel("$p(t)$")

# logit transform of data:
axs[1].errorbar(X_data, Y_data_transf, yerr=2*Y_data_transf_err, **plotstyle, label=f"{v}")

# transform of logistic fit of raw data to best line of transformed data.
axs[1].plot(X_data, logit(logistic_model(X_data_num, *popt_logis)), label=f"transform of Logistic fit")

# linear fit of transformed data
#label_str = f"linear model $\\alpha x + \\beta$; $\\alpha=${round(alpha, 2)}, $\\beta=${round(beta, 2)}"
label_str = f"linear reg."
axs[1].plot(X_data, lin_model(X_data_num, *popt_lin), linestyle="--", label=label_str)

# TODO: Make residual plots!
axs[1].tick_params(axis='x', labelrotation=90)
axs[1].set_ylabel("$\\ln \\left( \\frac{p(t)}{1-p(t)} \\right)$")

## Draw weekends
firstSunday = np.datetime64('2021-05-30') + np.timedelta64(7,'D')
numWeeks = 10
for k in range(0,numWeeks):
    curSunday = firstSunday + np.timedelta64(7*k,'D')
    axs[0].axvspan(curSunday-np.timedelta64(1,'D')-np.timedelta64(12,'h'),curSunday+np.timedelta64(12,'h'),facecolor='lightgrey',zorder=-1)
    axs[1].axvspan(curSunday-np.timedelta64(1,'D')-np.timedelta64(12,'h'),curSunday+np.timedelta64(12,'h'),facecolor='lightgrey',zorder=-1)
axs[0].axvspan(curSunday-np.timedelta64(1,'D')-np.timedelta64(12,'h'),curSunday+np.timedelta64(12,'h'),facecolor='lightgrey',label='Weekend')
axs[1].axvspan(curSunday-np.timedelta64(1,'D')-np.timedelta64(12,'h'),curSunday+np.timedelta64(12,'h'),facecolor='lightgrey',label='Weekend')

axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%d\n%B'))


leg_dict = {} #{"bbox_to_anchor":(1.05, 0.6)}
axs[0].legend(**leg_dict)
axs[1].legend(**leg_dict)

fig.suptitle("DK data")

saveplt(f"data_fits_DK_daily")
        
#num_seq_df # .loc[(num_seq_df.country=="Denmark")&(num_seq_df.variant=="B.1.617.2")]
```
