```python
# Flip dataframe to be like df1
dfdeaths = pd.DataFrame()
dfdeaths_once = pd.DataFrame()
dfdeaths_both = pd.DataFrame()
dfdeaths_full = pd.DataFrame()
dfdeaths_reva = pd.DataFrame()
dfdeaths_none = pd.DataFrame()

weekNames = dfmain.Ugenummer
dfdeaths['Ugenummer'] = weekNames
dfdeaths_none['Ugenummer'] = weekNames
dfdeaths_once['Ugenummer'] = weekNames
dfdeaths_both['Ugenummer'] = weekNames
dfdeaths_full['Ugenummer'] = weekNames
dfdeaths_reva['Ugenummer'] = weekNames

for k in range(0,len(dfcount)):
    curRow = dfcount.loc[k]
    curAge = dfcount.loc[k,'Aldersgruppe']
    # print(curAge)
    
    curNone = []
    curOnce = []
    curBoth = []
    curFull = []
    curReva = []
    for curW in weekNames:
        # print(curW)
        
        thisNone = curRow[curW+'_Ingen vaccination']
        curNone.append(thisNone)
        
        thisOnce = curRow[curW+'_Første vaccination']
        curOnce.append(thisOnce)
        
        thisBoth = curRow[curW+'_Anden vaccination']
        curBoth.append(thisBoth)
        
        thisFull = curRow[curW+'_Fuld effekt efter primært forløb']
        curFull.append(thisFull)
        
        thisReva = curRow[curW+'_Fuld effekt efter revaccination']
        curReva.append(thisReva)
    
    # Add to dataframe
    dfdeaths[curAge+'_Ingen vaccination'] = curNone
    dfdeaths[curAge+'_Første vaccination'] = curOnce
    dfdeaths[curAge+'_Anden vaccination'] = curBoth 
    dfdeaths[curAge+'_Fuld vaccineeffekt'] = curFull
    dfdeaths[curAge+'_Fuld effekt efter revaccination'] = curReva
    
    # Add to smaller dataframes as well
    dfdeaths_none[curAge] = curNone
    dfdeaths_once[curAge] = curOnce
    dfdeaths_both[curAge] = curBoth 
    dfdeaths_full[curAge] = curFull
    dfdeaths_reva[curAge] = curReva
    

# dfdeaths
```
