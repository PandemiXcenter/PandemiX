Den 28 marts blev vaccinations filerne helt ændret. 
Under findes en hurtig gennemgang af de nye filer.


## Vaccine_alder_koen_XXXX

Øjebliksbillede over hvormange stik hele befolkningen har fået, 1 2 eller 3, inkluderer også gruppe størrelserne.
Grupperne er Køn og aldersgrupper.

"Vaccine_region_alder_koen_XXX" Har 4 sufix i to grupper.
En gruppe der laven en alders opdeling på 10 års intervaller og en (dem der slutter på "XXinstage") In stages har en ujævn aldersgruppering.
Den anden gruppe er "vaccXX" og "repXX". vacc er data fra vaccinations datoen og rep re data fra indeberetningsdatoen.

## Vaccine_dato

Tidsserie (fra 2020-08-12) af daglige vaccinations stik samt stik nummer. Inkluderer også befolknings størrelsen (den er dog konstant, men den bruges til at udregne % dækning)

## Vaccine_dato_XXXX

Som "Vaccine_dato" men med en ekstra opløsning, der findes for region og kommune. 
Som i Vaccine_alder_koen_XXXX findes samme 4 sufix for Vaccine_dato_region_alder_XXXX. Disse filer indeholder alder og regions dimentioner for tidsserien. 

## Vaccine_kommune
Øjebliksbillede over Antal stik og dækning i danske kommuner (samme information indeholdes vel også i Vaccine_dato_XX ?)

## Vaccine_noegletal_kommune
Øjebliksbillede over antal stik, indeholder også ændring af stik grupper fra dagen før. Har alders og kommune opløsning

## Vaccine_noegletal_region
Samme som Vaccine_noegletal_kommune bare med Regioner.

## Vaccine_region
Øjebliksbillede af stik i regionerne (samme information er i Vaccine_noegletal_region)

## Vaccine_region_XXXX
Øjebliksbillede af stik i regionerne men med samme suffix opdeling som Vaccine_alder_koen_XXXX og Vaccine_dato_XXXX. Igen har den alder og kønsopdeling, samt totale gruppestørrelser.


## Vaccine_type_region
Øjebliksbillede af hvilken Vaccine der er givet i regionerne, for 1. 2. og 3. stik.

## Vaccine_uge_alder_XXXX
Tidsserie i uge opdeling (fra 2020-U50) med aldersopløsning, og de 4 suffix.
