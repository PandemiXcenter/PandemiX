# Viggos model

```python

# 'beta * S * I' : ['S','I'],

modelMeta = ({
     'beta * S * I' : ['S','I'],
     'beta*(np.sin((t+150)/365.2)+1) * S * I' : ['S','I'],
     'gamma * I' : ['I','S']
},
['S','I'],['beta','gamma'])
```

```python
curSch = pmf.Scheme(modelMeta,{'S':0.9,'I':0.1},{'beta':2/7,'gamma':1/47},0,5000)

curSch.simulate() 
```

```python

plt.figure(figsize=(5,2)) 
plt.plot(curSch.result.t,curSch.result.y.transpose())
```

```python

modelMeta = ({
     'beta_I * S * I' : ['S','I'],
     'beta_Y * S * Y' : ['S','Y'],
     'nu_I * I' : ['I',None],
     'nu_Y * Y' : ['Y',None]
},
['S','I','Y'],['beta_I','beta_Y','nu_I','nu_Y'])

varsInit = {
     'S' : 0.99,
     'I' : 0.0099,
     'Y' : 0.0001
}

parsInit = {
     'beta_I' : 1.2/7,
     'beta_Y' : 2/7,
     'nu_I' : 1/7,
     'nu_Y' : 1/7
}

curSch = pmf.Scheme(modelMeta,varsInit,parsInit,0,100)

curSch.simulate() 
```

```python

fig,(ax1,ax2) = plt.subplots(2,1,figsize=(6,4)) 
ax1.plot(curSch.result.t,curSch.result.I)
ax1.plot(curSch.result.t,curSch.result.Y)
# plt.plot(curSch.result.t,curSch.result.y.transpose())
ax2.plot(curSch.result.t,curSch.result.Y/(curSch.result.Y+curSch.result.I))
```
